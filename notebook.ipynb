## Exploratory Data Analysis (EDA)

### Explore ...

```python
# Konversi tanggal transaksi
orders_df["order_purchase_timestamp"] = pd.to_datetime(orders_df["order_purchase_timestamp"])

# Menghitung Recency (hari sejak transaksi terakhir)
latest_date = orders_df["order_purchase_timestamp"].max()
rfm = orders_df.groupby("customer_id").agg(
    Recency=("order_purchase_timestamp", lambda x: (latest_date - x.max()).days),
    Frequency=("order_id", "count")
).reset_index()

# Menghitung Monetary (Total uang yang dibelanjakan per pelanggan)
monetary = order_items_df.groupby("order_id")["price"].sum().reset_index()
rfm = rfm.merge(orders_df[["customer_id", "order_id"]], on="customer_id", how="left").merge(monetary, on="order_id", how="left")

# Kelompokkan ulang berdasarkan customer_id
rfm = rfm.groupby("customer_id").agg(
    Recency=("Recency", "min"),
    Frequency=("Frequency", "max"),
    Monetary=("price", "sum")
).reset_index()

# Jumlah pelanggan per kota
customer_city_counts = customers_df["customer_city"].value_counts().reset_index()
customer_city_counts.columns = ["customer_city", "count"]

# Jumlah pelanggan per negara bagian
customer_state_counts = customers_df["customer_state"].value_counts().reset_index()
customer_state_counts.columns = ["customer_state", "count"]
```

**Insight:**
- Kode yang telah dibuat mencakup berbagai jenis Exploratory Data Analysis (EDA), yaitu EDA Numerikal, EDA Kategorikal, EDA Univariate, dan EDA Bivariate.
  - EDA Numerikal diterapkan dalam analisis Recency, Frequency, dan Monetary (RFM Analysis), di mana setiap metrik berupa data kuantitatif yang dihitung berdasarkan transaksi pelanggan.
  - EDA Kategorikal digunakan dalam eksplorasi data lokasi pelanggan, seperti jumlah pelanggan per kota dan negara bagian, yang merupakan variabel kategori.
  - EDA Univariate dilakukan dengan menganalisis distribusi masing-masing variabel secara terpisah, seperti distribusi pelanggan berdasarkan lokasi atau nilai transaksi pelanggan.
  - EDA Bivariate diterapkan dalam analisis hubungan antara dua variabel, misalnya korelasi antara jumlah transaksi (Frequency) dengan total pembelanjaan (Monetary).
- Pola Pembelian Pelanggan Berdasarkan RFM Analysis
RFM (Recency, Frequency, Monetary) membantu dalam mengidentifikasi kelompok pelanggan berdasarkan perilaku pembelian mereka. Dari kode yang dieksekusi:

## Analisis Lanjutan (Opsional)

```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df_orders = pd.read_csv("https://raw.githubusercontent.com/acah21/dataset/refs/heads/main/orders_dataset.csv")

# Pastikan format tanggal benar
df_orders['order_purchase_timestamp'] = pd.to_datetime(df_orders['order_purchase_timestamp'])

# Ambil tanggal terakhir dalam dataset untuk perhitungan Recency
snapshot_date = df_orders['order_purchase_timestamp'].max() + dt.timedelta(days=1)

# Hitung Recency dan Frequency
df_rfm = df_orders.groupby('customer_id').agg(
    Recency=('order_purchase_timestamp', lambda x: (snapshot_date - x.max()).days),  # Hari sejak transaksi terakhir
    Frequency=('order_id', 'count')  # Jumlah transaksi
).reset_index()

# Cek apakah ada kolom "price" untuk Monetary
if 'price' in df_orders.columns:
    df_monetary = df_orders.groupby('customer_id')['price'].sum().reset_index()
    df_monetary.rename(columns={'price': 'Monetary'}, inplace=True)
else:
    # Jika tidak ada price, gunakan jumlah transaksi sebagai proxy Monetary
    df_monetary = df_rfm[['customer_id', 'Frequency']].copy()
    df_monetary.rename(columns={'Frequency': 'Monetary'}, inplace=True)

# Gabungkan Recency, Frequency, dan Monetary
df_rfm = df_rfm.merge(df_monetary, on='customer_id', how='left')

df_rfm['R_Score'] = pd.qcut(df_rfm['Recency'], 4, labels=[4, 3, 2, 1], duplicates='drop')
df_rfm['F_Score'] = pd.qcut(df_rfm['Frequency'].rank(method="first"), 4, labels=[1, 2, 3, 4], duplicates='drop')

# Fix untuk Monetary (gunakan pd.cut jika perlu)
try:
    df_rfm['M_Score'] = pd.qcut(df_rfm['Monetary'], 4, labels=[1, 2, 3, 4], duplicates='drop')
except ValueError:
    df_rfm['M_Score'] = pd.cut(df_rfm['Monetary'], bins=4, labels=[1, 2, 3, 4], right=False)

# Membuat segmentasi RFM
df_rfm['RFM_Score'] = df_rfm['R_Score'].astype(str) + df_rfm['F_Score'].astype(str) + df_rfm['M_Score'].astype(str)

# Menentukan kategori pelanggan berdasarkan skor RFM
def categorize_rfm(df):
    if df['RFM_Score'] in ['444', '443', '433', '434', '344']:
        return 'Best Customers'
    elif df['RFM_Score'] in ['411', '422', '412']:
        return 'Loyal Customers'
    elif df['RFM_Score'] in ['211', '311', '411']:
        return 'New Customers'
    elif df['RFM_Score'] in ['144', '134', '124']:
        return 'Churned Customers'
    else:
        return 'Regular Customers'

df_rfm['Customer_Segment'] = df_rfm.apply(categorize_rfm, axis=1)

# Menyimpan hasil RFM ke CSV
df_rfm.to_csv("rfm_analysis_results.csv", index=False)

# ===  VISUALISASI  === #

plt.figure(figsize=(15, 5))

# Distribusi Recency
plt.subplot(1, 3, 1)
sns.histplot(df_rfm['Recency'], bins=20, kde=True, color='blue')
plt.xlabel("Recency (Hari)")
plt.ylabel("Jumlah Pelanggan")
plt.title("Distribusi Recency")

# Distribusi Frequency
plt.subplot(1, 3, 2)
sns.histplot(df_rfm['Frequency'], bins=20, kde=True, color='green')
plt.xlabel("Frequency (Jumlah Transaksi)")
plt.ylabel("Jumlah Pelanggan")
plt.title("Distribusi Frequency")

# Distribusi Monetary
plt.subplot(1, 3, 3)
sns.histplot(df_rfm['Monetary'], bins=20, kde=True, color='red')
plt.xlabel("Monetary (Total Belanja)")
plt.ylabel("Jumlah Pelanggan")
plt.title("Distribusi Monetary")

plt.tight_layout()
plt.show()

# Visualisasi Segmentasi Pelanggan (PERBAIKAN)
plt.figure(figsize=(8, 5))
sns.countplot(y=df_rfm['Customer_Segment'],
              order=df_rfm['Customer_Segment'].value_counts().index,
              hue=df_rfm['Customer_Segment'],  # Menetapkan hue agar tidak kena warning
              palette="viridis",
              legend=False)  # Menghilangkan legend agar tidak redundant
plt.xlabel("Jumlah Pelanggan")
plt.ylabel("Segmentasi Pelanggan")
plt.title("Segmentasi Pelanggan Berdasarkan RFM")
plt.show()
```

## Conclusion

1. Pola Pembelian Pelanggan Berdasarkan Recency, Frequency, dan Monetary (RFM Analysis)

- Dari hasil analisis Recency, Frequency, dan Monetary (RFM), ditemukan beberapa pola utama dalam perilaku pembelian pelanggan:

  - Distribusi Recency menunjukkan bahwa sebagian besar pelanggan melakukan transaksi dalam rentang waktu tertentu, tetapi ada juga segmen pelanggan yang sudah lama tidak bertransaksi. Ini mengindikasikan bahwa terdapat pelanggan aktif dan pelanggan yang mulai tidak aktif (churn).
  - Distribusi Frequency menunjukkan bahwa sebagian besar pelanggan hanya melakukan sedikit transaksi dalam periode tertentu, sementara hanya sebagian kecil pelanggan yang bertransaksi berkali-kali. Hal ini menegaskan bahwa pelanggan setia (loyal customers) merupakan bagian kecil dari keseluruhan pelanggan.
  - Distribusi Monetary menunjukkan bahwa sebagian besar pelanggan memiliki pengeluaran yang rendah hingga sedang, sementara hanya sedikit pelanggan yang memiliki pengeluaran tinggi. Ini mengindikasikan adanya pelanggan VIP atau high-value customers yang bisa menjadi target program loyalitas.

- Berdasarkan segmentasi RFM Score, pelanggan dikategorikan menjadi beberapa kelompok:

  - Best Customers : Pelanggan yang sering berbelanja, baru-baru ini melakukan transaksi, dan memiliki total pengeluaran tinggi. Kelompok ini sebaiknya diberikan program loyalitas, diskon eksklusif, atau akses prioritas ke produk baru.
  - Loyal Customers : Pelanggan dengan frekuensi belanja tinggi tetapi mungkin tidak memiliki nilai transaksi sebesar best customers. Program keanggotaan atau rewards bisa diberikan untuk mempertahankan mereka.
  - New Customers : Pelanggan yang baru pertama kali bertransaksi dan memiliki potensi untuk menjadi pelanggan tetap jika diberikan pengalaman belanja yang positif.
  - Churned Customers : Pelanggan yang sudah lama tidak melakukan transaksi. Strategi pemasaran ulang seperti email promosi atau diskon khusus dapat digunakan untuk menarik mereka kembali.
  - Regular Customers : Pelanggan dengan pola belanja stabil namun tidak menonjol dalam aspek recency, frequency, maupun monetary.

- Dari hasil ini, dapat disimpulkan bahwa strategi pemasaran berbasis segmentasi pelanggan dapat meningkatkan retensi pelanggan dan nilai transaksi secara keseluruhan.

2. Distribusi Geografis Pelanggan Berdasarkan Kota dan Negara Bagian
- Analisis geospasial terhadap distribusi pelanggan berdasarkan kota dan negara bagian menunjukkan pola berikut:
  - Sebagian besar pelanggan berasal dari kota-kota besar, yang mengindikasikan bahwa layanan e-commerce lebih banyak digunakan di wilayah urban dengan infrastruktur yang lebih mendukung.
  - Negara bagian dengan jumlah pelanggan tertinggi merupakan area dengan potensi pasar terbesar. Strategi pemasaran dan distribusi produk dapat lebih difokuskan di wilayah ini untuk meningkatkan efisiensi bisnis.
  - daerah dengan jumlah pelanggan lebih sedikit mungkin menghadapi kendala dalam akses ke layanan e-commerce, seperti keterbatasan pengiriman atau kurangnya kesadaran akan layanan yang tersedia. Wilayah ini dapat dijadikan target ekspansi dengan strategi seperti promosi ongkir gratis, diskon wilayah tertentu, atau kampanye pemasaran berbasis lokal.
  - Pemetaan geospasial juga menunjukkan bahwa beberapa wilayah memiliki nilai transaksi per pelanggan yang lebih tinggi dibandingkan daerah lain. Ini dapat digunakan untuk menentukan strategi harga, produk yang lebih diminati, serta optimasi rantai pasokan.

- Dengan memahami pola geografis ini, bisnis dapat menyesuaikan strategi pemasaran dan logistik untuk meningkatkan kepuasan pelanggan serta memperluas jangkauan pasar di wilayah yang masih memiliki potensi pertumbuhan.
